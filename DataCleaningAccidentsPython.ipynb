```python
df= pd.read_csv("US_Accidents_Dec21_updated.csv")
```

```python
df = df.drop('Number', axis=1)
```

```python
df = df.drop('Wind_Chill(F)', axis=1)
df = df.drop('Precipitation(in)', axis=1)
```

```python
%%time
plt.figure(figsize=(10,6))
sns.heatmap(df.isna(), cbar=False, cmap='viridis', yticklabels=False)
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df.dropna(axis=0, how='any', inplace=True)
```

```python
df.shape
```

```python
%%time
plt.figure(figsize=(10,6))
sns.heatmap(df.isna(), cbar=False, cmap='viridis', yticklabels=False)
```

```python
df['Wind_Direction'].unique()
```

```python
df.loc[df["Wind_Direction"] == "North", "Wind_Direction"] = 'N'
df.loc[df["Wind_Direction"] == "South", "Wind_Direction"] = 'S'
df.loc[df["Wind_Direction"] == "West", "Wind_Direction"] = 'W'
df.loc[df["Wind_Direction"] == "East", "Wind_Direction"] = 'E'
df.loc[df["Wind_Direction"] == "WSW", "Wind_Direction"] = 'SW'
df.loc[df["Wind_Direction"] == "WNW", "Wind_Direction"] = 'NW'
df.loc[df["Wind_Direction"] == "NNW", "Wind_Direction"] = 'NW'
df.loc[df["Wind_Direction"] == "SSE", "Wind_Direction"] = 'SE'
df.loc[df["Wind_Direction"] == "SSW", "Wind_Direction"] = 'SW'
df.loc[df["Wind_Direction"] == "ESE", "Wind_Direction"] = 'SE'
df.loc[df["Wind_Direction"] == "Variable", "Wind_Direction"] = 'VAR'
df.loc[df["Wind_Direction"] == "NNE", "Wind_Direction"] = 'N'
df.loc[df["Wind_Direction"] == "ENE", "Wind_Direction"] = 'NE'
```

```python
df['Wind_Direction'].unique()
```

```python
df.to_csv('US_Accidents_Dec21_updated1.csv', index=False)
```
